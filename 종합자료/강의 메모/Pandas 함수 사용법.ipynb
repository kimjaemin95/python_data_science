```python
data = [
    {"email":"dss@gmail.com","name":"dss","id":1},
    {"email":"fcamp@daum.net","name":"fcamp","id":2},
    {"email":"andy@naver.com","name":"andy","id":3},
]
```

```python
df = pd.DataFrame(data)
df
```

### apply() 함수 사용법
- map 함수와 비슷

```python
# email 컬럼에서 메일의 도메인만 가져와서 새로운 도메인 컬럼을 생성하고 싶을 때
```

```python
def domain(email):
    return email.split("@")[1].split(".")[0]
```

```python
# df['email'] 은 Series 데이터 이다.
df['domain'] = df['email'].apply(domain)
df
```

```python
df['domain2'] = df['email'].apply(lambda x : x.split("@")[1].split(".")[0])
df
```

### append() 함수 사용법

### reset_index 함수 사용법

```python
# reset_index 함수로 인덱스 재정렬
# drop : reset_index하면서 생긴 index 컬럼 DROP 여부
# inplace : reset_index 후 DF에 데이터 값을 저장
df3.reset_index(drop=True, inplace=True)
df3.tail(3)
```

### concat() 함수 사용법
- axis 0일 때 : 세로로 합치기
- axis 1일 때 : 가로로 합치기

```python
df3 = pd.concat([df1, df2], axis = 0).reset_index(drop=True)
df3
```

```python
# join='outer'로 하면 인덱스가 부족한 df1의 값들이 Nan으로 출력 됌
pd.concat([df3, df1], axis=1, join='inner')
```

### groupby
- 특정 컬럼의 중복 되는 데이터를 합쳐서 새로운 데이터 프레임을 만드는 방법

```python
df.groupby("Name").size().reset_index(name="count")
```

```python
# Series 데이터
df.groupby("Name").size()
```

```python
df_gp = df.groupby("Name").agg(
    name_count = ("Name", "count")
).reset_index()
df_gp
```

### sort_values() 함수 
- 설정한 컬럼으로 데이터 프레임을 정렬

```python
df_gp.sort_values(["name_count"])
```

```python
df_gp.sort_values(["name_count"], ascending=False)
```

```python
df_gp.sort_values(["name_count"], ascending=False, inplace=True)
df_gp
```

```python
df_gp.reset_index(drop=True, inplace=True)
df_gp
```
