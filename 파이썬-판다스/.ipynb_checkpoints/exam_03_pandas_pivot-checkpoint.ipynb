### Pandas Pivot
- 데이터 프레임에서 컬럼 데이터에서 index, column, value 를 선택해서 데이터 프레임을 만든는 방법
- `df.pivot(index, columns, values)` : groupby 후 pivot을 실행
- `df.pivot_table(values, index, columns, aggfunc)` : groupby 없이 pivot_table 실행
