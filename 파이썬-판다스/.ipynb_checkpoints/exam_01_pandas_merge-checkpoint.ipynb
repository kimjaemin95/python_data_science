### quiz
- makedata 모듈을 이용해서 데이터 프레임 만들기
- user_df
    - 8명의 데이터가 있는 데이터 프레임을 만드세요.
    - UserID : 1 ~ 8 
    - Name : makedata.get_name()
    - Age : makedata.get_age()
    - 중복되는 Name 값이 없도록

```python
mk.get_age(), mk.get_name()
```

```python
# 딕셔너리의 리스트
datas = {}
datas["UserID"] = list(range(1,9))
datas["Age"] = [mk.get_age() for _ in range(8)]

names = []
while True:
    name = mk.get_name()
    if name not in names:
        names.append(name)
    if len(names) >= 8:
        break
datas["Name"] = names        


user_df = pd.DataFrame(datas)
```

```python
user_df
```

```python
# 리스트 딕셔너리
user_df = pd.DataFrame(columns=["UserID", "Age", "Name"])

for idx in range(1,9):
    
    name = mk.get_name()
    while name in list(user_df["Name"]):
        name = mk.get_name()
        
    data = {"UserID":idx, "Age":mk.get_age(), "Name":name}
    user_df.loc[len(user_df)] = data
    
user_df
```

### quiz
- money_df 만들기
- 15개의 데이터가 있습니다.
- columns :
        - ID : 1~8
        - Money : 1,000원 단위 1,000원 ~ 20,000원까지

```python
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
money_df = pd.DataFrame(columns=["ID", "Money"])
money_df
```

```python
for _ in range(15):
    money_df.loc[len(money_df)] = {
        "ID":np.random.randint(1,9),
        "Money":np.random.randint(1,21)*1000
    }

# 컬럼데이터에서 Unique 값 확인
ids = money_df["ID"].unique()
ids.sort()
ids
```

```python
money_df
```

```python
user_df
```

```python
df_merge = user_df.merge(money_df,left_on=user_df["UserID"], right_on=money_df["ID"], how='left')
df_merge.fillna(0, inplace=True)
```

```python
df_gp = df_merge.groupby("Name").agg(
    SumMoney = ("Money", "sum"),
    AvgMoney = ("Money", "mean"),
    MaxMoney = ("Money", "max"),
    MinMoney = ("Money", "min"),
    UserCount = ("Name", "count")
).reset_index()
```
