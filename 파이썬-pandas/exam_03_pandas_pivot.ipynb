### 성별, 좌석등급에 따른 데이터의 수

```python
df_train.columns
```

```python
df_1 = df_train.groupby(["Sex", "Pclass"]).size().reset_index(name="counts")
df_1
```

### pivot 함수

```python
result = df_1.pivot(index='Sex', columns='Pclass', values='counts')
result
```

### pivot_table 함수

```python
df_train["counts"] = 1
df_train
```

```python
'''
Signature:
df_train.pivot_table(
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
) -> 'DataFrame'
'''
result = df_train.pivot_table(values="counts", index=["Pclass"], columns=["Survived"], aggfunc=np.sum)
result
```

```python
result['total'] = result[0] + result[1]
result
```

```python
result.loc["total"] = result.loc[1] + result.loc[2] + result.loc[3]
result
```
